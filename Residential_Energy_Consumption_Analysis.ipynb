```python
#Import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
from scipy.stats import pearsonr
```

```python
#Pull up Residential Energy Consumption throughout 50 states (including USA totals) dataframe from CSV file
df = pd.read_csv (r'Residential_Energy_Consumption.csv')
print (df)
```

```python
#Summary Statistics
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the energy consumption by state
#(including USA totals) per year (2000-2015)

mean = df.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']
median = df.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
var = df.groupby('Drug Regimen').var()['Tumor Volume (mm3)']
drug_std = df.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
SEM = df.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']

summary_df = pd.DataFrame({"Mean": drug_mean, "Median": drug_median, "Variance": drug_var, "Standard Deviation": drug_std,
                           "Standard Error": drug_SEM})
summary_df
```
