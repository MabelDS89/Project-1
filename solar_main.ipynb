```python
state_id = "SEDS.SOR7P.NC.A"
series_url = f"http://api.eia.gov/series/?api_key={api_key}&series_id="
query_url = (series_url + state_id)

index2 = np.arange(0,19,1)
year_list = []
consumption_list = []

for x in range(len(index2)):
    year = requests.get(query_url).json()['series'][0]['data'][0:19][index2[x]][0]
    consumption = requests.get(query_url).json()['series'][0]['data'][0:19][index2[x]][1]
    year_list.append(year)
    consumption_list.append(consumption)

print(year_list)
print(consumption_list)
```
