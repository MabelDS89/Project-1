```python
#Import dependencies
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress
from scipy.stats import pearsonr
```

```python
#Pull up Residential Energy Consumption throughout 50 states (including USA totals) dataframe from CSV file
df_res = pd.read_csv (r'Residential_Energy_Consumption_Updated.csv')
print (df_res)
```

```python
#Pull up Commercial and Industrial Energy Consumption throughout 50 states (including USA totals) dataframe
# from CSV file
df_com = pd.read_csv (r'Resources/Commercial_Energy_Consumption.csv')
print (df_com)
```

```python
#Pull up Commercial and Industrial Energy Consumption throughout 50 states (including USA totals) dataframe
# from CSV file
df_ind = pd.read_csv (r'Resources/Industrial_Energy_Consumption.csv')
print (df_ind)
```

```python
#Merge 3 DataFrames together
df_merged = pd.merge(df_res, df_com, on="State", suffixes=("_Res", "_Com"))
df_merged.head()
```

```python
df_merged_final = pd.merge(df_merged, df_ind, on="State", suffixes=("", "_Ind"))
# df_merged_final.set_option("max_columns", none)
list(df_merged_final.columns)
```
